### Part 1: Theme Park Data Analysis

```python
from __future__ import print_function, division
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import axes
from scipy import integrate
```

```python
# Load the velocity and time data from droptower_vdata.txt
t, v = np.loadtxt('droptower_vdata.txt', skiprows=5, unpack=True)
```

```python
# Calculate position:
y = 30 + integrate.cumtrapz(v, t, initial=0)
```

```python
# Calculate acceleration:
a = np.diff(v)
tnew = np.linspace(0.5, 9.5, 10)
```

```python
# Plotting position, velocity, and accelerations
plt.scatter(t,y, color="purple", marker="o")
plt.title("Position vs. Time")
plt.xlabel("Time (s)")
plt.ylabel("Position (m)")
plt.show()

plt.scatter(t,v, color="blue", marker="o")
plt.title("Velocity vs. Time")
plt.xlabel("Time (s)")
plt.ylabel("Velocity (m/s)")
plt.show()

plt.scatter(tnew,a, color="green", marker="o")
plt.title("Acceleration vs. Time")
plt.xlabel("Time (s)")
plt.ylabel("Acceleration (m/s^2)")
plt.show()
```

### Part 2: Drop Tower Continued

```python
# Now we get the positive accelerations:
for i in range(0, a.size):
    if a[i] < 0:
        a[i] = 0
# Amount of pos elements
count = 0;
for i in range(0, a.size):
    if a[i] > 0:
        count = count + 1
# Average them:
avg_acc = np.sum(a) / count
```

```python
print("Average acceleration: ", avg_acc, " m/s^2")
```
